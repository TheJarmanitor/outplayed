```python
import pandas as pd

df = pd.read_csv("../data/player_stats.csv")
df["Percent"] = df["Percent"].fillna(0.0)
df = df.dropna(axis=1)
```

```python
df.head(30)
```

```python
df = df.drop(columns=["Season", "Percent"])
df = df.set_index(["Year", "Rank", "Player", "Club", "Nationality"])
```

```python
df.sum()
```

```python
X_train = df.query("Year < 23")
X_test = df.query("Year >= 23")

y_train = X_train.pop("Share")
y_test = X_test.pop("Share")
```

```python
import statsmodels.api as sm
```

```python
glm1 =  sm.GLM(endog=y_train, exog=sm.add_constant(X_train), family=sm.families.Gaussian())
# stats23 = sm.add_constant(X_train)
results = glm1.fit()
results.summary()

'''
Only significant variables: goals, Assists, Appearances in Champions, Serie A & Clubs World Cup

'''
```

```python
limit = X_train.columns.get_loc("UEFA_Super_Cup_appearances")

X_train = X_train.iloc[:, :limit]

glm2 =  sm.GLM(endog=y_train, exog=sm.add_constant(X_train), family=sm.families.Gaussian())
# stats23 = sm.add_constant(X_train)
results = glm2.fit()
results.summary()

# '''
# Only significant variables: goals, Assists, Appearances in Champions & Clubs World Cup, international and domestic competitions

# '''

```

```python
X_train = X_train.drop(columns=df.columns[[9,10,13,14,15,16,17,18,19,20]])

glm3 =  sm.GLM(endog=y_train, exog=sm.add_constant(X_train), family=sm.families.Gaussian())
# stats23 = sm.add_constant(X_train)
results = glm3.fit()
results.summary()

# '''
# Only significant variables: goals, Assists, Appearances in Champions & Clubs World Cup, international and domestic competitions

# '''
```
