```python
## Player trophies 

players = pd.read_csv("..\data\player_stats_new.csv")
players = pd.DataFrame(players)

#christiano_08_09["UEFA-Cup_Qualifikation_appearances"]
print(players.loc[10])
```

```python
season_08_09 = players.loc[players["Year"]==8]
#christiano_08_09 = season_08_09.loc[season_08_09["Player"]=="Cristiano Ronaldo"]
y = season_08_09["Share"].values
season_08_09 = season_08_09.drop(['Year', "Share", 'Rank', "Voted", "Nationality", "Points", "P1", "P2", 
                                  "P3", "P4", "P5", "Votes", "RankPts", "Percent", "Votes", "Season"], axis=1)
print(season_08_09)
```

```python
important_stats = season_08_09[["Player", "Club","Goals", "Assists","Yellow cards", 
                                "Minutes played", "Goals conceded", "Clean sheets",
                               "Champions_League_appearances", "UEFA_Cup_appearances"]]
#print(y)
display(important_stats)
```

```python
#!pip install bayesian-bootstrap
list(important_stats.columns)[2:]
```

```python
plt.figure(figsize=(10,10))
```

```python
from bayesian_bootstrap import mean, highest_density_interval
# Set the figure size for the subplots
# Set the figure size for the subplots
fig, axs = plt.subplots(len(list(important_stats.columns)[2:]), figsize=(12, 18), sharex=False)  # Change the size as needed

for j, i in enumerate(list(important_stats.columns)[2:]):
    posterior_samples = mean(important_stats[i], 10000)
    l, r = highest_density_interval(posterior_samples)

    sns.distplot(posterior_samples, label='Bayesian Bootstrap Samples', ax=axs[j])
    axs[j].plot([l, r], [0, 0], linewidth=5.0, marker='o', label='95% HDI')
    axs[j].set_title('Bayesian Bootstrap of mean for ' + i)  # Set individual subplot titles

# Adjust spacing between subplots
plt.subplots_adjust(hspace=1.5, wspace=10.5)  # Change the values to increase or decrease the spacing

plt.show()
```
